# You are given a text file named input.txt containing a list of words, one word per line. Your task is to create a Python program that reads the contents of input.txt, processes the words, and writes the result to an output file named output.txt. a. The program should perform the following operations: i. Read the words from input.txt. ii. For each word in the input file, calculate the length of the word and store it in a dictionary where the word is the key, and the length is the value. iii. Write the word-length dictionary to output.txt in the following format: Word1: Length1 Word2: Length2 iv. Close both input and output files properly. v. Write Python code to accomplish this task. Ensure proper error handling for fileÂ operations.

```python
def read_words(file_path):
    try:
        with open(file_path, 'r') as file:
            words = file.read().split()
        return words
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

file_path = 'one.txt'
words_list = read_words(file_path)

print("Words from 'one.txt':")
for word in words_list:
    print(word)
```

```python
word_= {}
with open('one.txt', 'r') as file:
    content = file.read()
    words = content.split()
    
    for word in words:
        word = word.strip(".,!?()[]{}\"'")
        length = len(word)
        word_[word] = length

for word, length in word_.items():
    print(f'{word}: {length}')
```

```python
word_len = {}
with open('one.txt', 'r') as file:
  
    content = file.read()
    words = content.split()
    for word in words:
        word = word.strip(".,!?()[]{}\"'")
        length = len(word)
        word_len[word] = length

with open('output.txt', 'w') as output_file:
    for word, length in word_len.items():
        output_file.write(f'{word}: {length}\n')

print("Word-length dictionary has been written to output.txt")
```
